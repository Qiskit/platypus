### Exercise 1: Spinning Around the Bloch Sphere

In quantum computing, information is stored in qubits, which are quantum bits that can exist in a superposition of states rather than either just the state of 0 or the state of 1.

We use what is known as the Bloch Sphere to visualize individual qubit states and the rotations around qubits that occur when quantum gates act on qubits. The state of a qubit is represented as a point on the surface of a unit sphere, where the 0 state (zero state, denoted by |0>) points upwards in the positive z-direction (you can think of this as the north pole of the sphere) and the 1 state (one state, denoted by |1>) points downwards in the negative z-direction (the south pole). Any other point on the sphere represents some superposition of the one and zero states. 

When rotations are applied to a qubit, rotations with a positive value move anti-clockwise around the sphere.

In the exercises _Spinning around the Bloch Sphere_ and _Let's Create a Quantum Circuit!_, you created two qubits that each had a property called quantum superposition. Superposition allows a qubit to represent multiple values at once, which is important for quantum computing to provide larger computing power.

You can think of superposition as an equal likelihood that a qubit will either be measured as a 0 or 1, and the Hadamard gate creates this superposition. If you measured 100 different qubits encoded in the same superposition state, your results would have a mixture of around 50 zeroes and 50 ones.  

Quantum entanglement is the phenomenon in which two or more qubits can become correlated such that the state of one system is dependent on the other system. Superposition and entanglement can work together to allow for the significant processing power of quantum computers.
